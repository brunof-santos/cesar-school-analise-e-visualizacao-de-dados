# O que é Pokémon?

Os Pokémon são criaturas fictícias que habitam um mundo fictício chamado de Mundo Pokémon. Eles são os principais protagonistas da franquia de mídia Pokémon, que inclui jogos de vídeo, desenhos animados, filmes, card games e uma variedade de outros produtos.

Cada Pokémon é caracterizado por suas próprias individualidades, incluindo habilidades, status, egg groups e tipos:

- Habilidades: As habilidades dos Pokémon são características distintas que podem afetar o curso das batalhas ou interações no jogo. Algumas habilidades aumentam os atributos do Pokémon, outras alteram os efeitos dos movimentos ou têm efeitos passivos.
- Status: Os status dos Pokémon referem-se aos seus atributos básicos, como HP (pontos de vida), ataque, defesa, sp_attack (ataque especial), sp_defense (defesa especial) e velocidade. Cada Pokémon tem seus próprios valores de status, que podem variar amplamente entre as espécies.
- Egg Groups: Os egg groups definem com quais outros Pokémon um determinado Pokémon pode acasalar para produzir ovos. Esses grupos são usados para determinar a compatibilidade de reprodução entre Pokémon no jogo.
- Tipos: Os Pokémon têm um ou dois tipos, que determinam suas fraquezas e resistências a diferentes tipos de ataques. Existem 18 tipos diferentes no total, incluindo tipos como grama, fogo, água, elétrico, psíquico, dragão, entre outros. A combinação de tipos de um Pokémon pode afetar significativamente sua eficácia em batalha. Exemplo: Um Pokémon do tipo grama recebe dano dobrado de um ataque tipo fogo, mas recebe metade do dano de um pokémon tipo água. A combinação de dois tipos pode trazer imunidades diferentes e resistências de até 1/4 de dano, mas também receber 4 vezes o dano de determinados tipos.
- Características físicas: Altura e peso também podem influenciar nas batalhas, pois existem golpes e habilidades que levam em consideração essas características
- Porcentagem na natureza de machos e fêmeas: Alguns Pokémon têm diferenças de gênero em suas proporções de machos e fêmeas na natureza. Enquanto alguns Pokémon têm uma divisão de gênero de 50/50, outros podem ter uma proporção maior de machos ou fêmeas. Isso pode afetar a criação e a obtenção de Pokémon com determinados atributos ou movimentos, além da eficácia de movimentos que usam esse atributo com base.
- Pokémon sem gênero: Alguns Pokémon são considerados sem gênero e não têm distinção entre macho e fêmea. Isso geralmente se aplica a Pokémon lendários, míticos e alguns Pokémon semelhantes a objetos.

Essas características únicas contribuem para a diversidade e complexidade do mundo Pokémon, tornando cada espécie única e oferecendo uma ampla gama de estratégias e possibilidades no jogo.

Além disso existem características (também únicas a cada espécie ou até indivíduo) que dão mais informações sobre o contexto em que ele foi inserido

- Gerações: As gerações referem-se aos diferentes períodos de lançamento dos jogos principais da série Pokémon. Cada geração, geralmente, coincide com o lançamento de novos jogos principais da Nintendo. Cada geração introduz novos Pokémon, mecânicas de jogabilidade, regiões exploráveis e histórias. Atualmente a franquia possui 9 gerações.
- Lendário/Sub Lendário/Mítico: Esses termos são usados para descrever certos Pokémon que são considerados especiais ou únicos dentro do universo Pokémon. Os Pokémon lendários são poderosos e geralmente têm um papel significativo na narrativa do jogo. Os sub lendários são uma categoria menos restrita, mas ainda incluem Pokémon raros e poderosos, com participações em histórias secundárias. Os Pokémon míticos são extremamente raros e associados à lendas e mitos no universo, e geralmente só podem ser obtidos através de eventos especiais ou distribuição online.
- Catch Rate: A taxa de captura é uma medida da facilidade com que um jogador pode capturar um Pokémon selvagem durante uma batalha. Quanto mais alta a taxa de captura, mais fácil é capturar o Pokémon. A taxa de captura varia de acordo com o Pokémon e pode ser influenciada por fatores como o status de saúde do Pokémon selvagem, o uso de Poké Balls específicas e outras condições.
- Base Experience: A experiência base é a quantidade de experiência que um jogador recebe ao derrotar um Pokémon selvagem. Cada Pokémon tem uma quantidade específica de experiência base associada a ele, que é usada para calcular a quantidade de experiência que os Pokémon do jogador recebem após uma batalha.
- Egg Cycles: Os ciclos de ovos representam o tempo necessário para que um ovo de Pokémon choque e produza um Pokémon recém-nascido. Cada espécie de Pokémon tem um número específico de ciclos de ovos necessários para chocar, que pode variar de algumas centenas a milhares de ciclos, dependendo da espécie. Um ciclo possui 255-257 passos (dependendo do jogo)

# Imagem com Top Maiores (Status a definir pelo usuário)

```python
# Definindo o status desejado para filtrar os Pokémon
# Pode ser 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'
status_desejado = 'speed'
```

```python
# Filtrando os Pokémon com base no status desejado
data = df_pokemon
top_5_status = data.nlargest(5, status_desejado)
```

```python
# Função para obter URL da imagem
def get_image_url(pokedex_number):
    number_str = str(pokedex_number).zfill(4)
    return f"https://pokejungle.net/sprites/normal/{number_str}.png"
```

```python
# Criando uma lista para armazenar as imagens e informações
images_info = []

for _, row in top_5_status.iterrows():
    name = row['name']
    status_value = row[status_desejado]
    url = get_image_url(row['pokedex_number'])

    response = requests.get(url)
    if response.status_code == 200:
        image = Image.open(BytesIO(response.content))
        images_info.append((name, status_value, image))
    else:
        print(f"Erro ao obter a imagem do Pokémon com Pokedex Number {row['pokedex_number']}")
```

```python
# Configurações da imagem final
image_width = 600
image_height = 800
font_size = 40
```

```python
# Criando uma nova imagem combinada
combined_image = Image.new('RGB', (image_width * 5, image_height), (255, 255, 255))

draw = ImageDraw.Draw(combined_image)
font = ImageFont.truetype("/content/cesar-school-analise-e-visualizacao-de-dados/arial.ttf", font_size)

for i, (name, status_value, image) in enumerate(images_info):
    # Colando a imagem  de cada poke na imagem combinada com uma borda de 5 pixels
    combined_image.paste(image, (i * image_width, 0))
    draw.rectangle([(i * image_width, 0), ((i + 1) * image_width, image_height)], outline="black", width=5)

    # Adicionando texto (nome e status)
    text = f"{name}\n{status_desejado.capitalize()}: {status_value}"
    text_width, text_height = draw.textsize(text, font=font)
    text_x = i * image_width + (image_width - text_width) // 2
    text_y = image_height - text_height - 20
    draw.text((text_x, text_y), text, font=font, fill="black")
```

```python
# Exibindo a imagem final usando matplotlib
plt.figure(figsize=(30, 20)) # Tamanho da Imagem
plt.imshow(combined_image)
plt.axis('off') # Tira os eixos
plt.show()
```

# Radar chart: média de cada status por tipo

```python
# Definindo as cores de cada tipo
cores_tipos = {
    'Grass': '#78C850', 'Fire': '#F08030', 'Water': '#6890F0', 'Electric': '#F8D030',
    'Ice': '#98D8D8', 'Steel': '#B8B8D0', 'Rock': '#B8A038', 'Ground': '#E0C068',
    'Bug': '#A8B820', 'Normal': '#A8A878', 'Dark': '#705848', 'Ghost': '#705898',
    'Psychic': '#F85888', 'Fairy': '#EE99AC', 'Dragon': '#7038F8', 'Flying': '#A890F0',
    'Fighting': '#C03028', 'Poison': '#A040A0'
}

tipos = list(cores_tipos.keys())
```

```python
# Função para criar dados do gráfico de radar
def dados_grafico_radar(tipo_desejado):
    # Filtrando o DataFrame pelo tipo da função
    df_tipo = df_pokemon[df_pokemon['type_all'].str.contains(tipo_desejado)]

    # Calculando a média dos status dos Pokémon do tipo da função
    status_cols = ['hp', 'sp_attack', 'sp_defense', 'speed', 'defense','attack']
    status_means = df_tipo[status_cols].mean()

    # Ordenando conforme a ordem do padrão dos jogos
    ordem = ['hp', 'sp_attack', 'sp_defense', 'speed', 'defense','attack']
    status_means = status_means[ordem]

    # Alterando nomes da legenda
    legenda = ['HP', 'Sp. Atk', 'Sp. Def', 'Speed', 'Defense','Attack']
    values = status_means.values
    num_vars = len(legenda)

    # Preparando dados para o gráfico
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    values = np.concatenate((values, [values[0]]))
    angles += angles[:1]

    # Rotacionando o gráfico em 90 graus para ficar no padrão dos jogos
    rotation_angle = np.deg2rad(90)
    angles = [(angle + rotation_angle) % (2 * np.pi) for angle in angles]

    return angles, values, legenda
```

```python
# Criando subplots
fig, axs = plt.subplots(6, 3, figsize=(18, 24), subplot_kw=dict(polar=True))

# Iterando sobre os tipos e criando cada gráfico
for i, tipo in enumerate(tipos):
    row = i // 3
    col = i % 3
    angles, values, legenda = dados_grafico_radar(tipo)

    ax = axs[row, col]
    cor = cores_tipos[tipo]

    ax.fill(angles, values, color=cor, alpha=0.25)
    ax.plot(angles, values, color=cor, linewidth=2)
    ax.set_yticks([0, 30, 60, 90, 120])
    ax.set_yticklabels(['0', '30', '60', '90', '120'])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(legenda)
    ax.set_title(tipo, size=12, color='black', y=1.1)

    # Ajustando o limite do eixo y para que o círculo maior seja 120
    ax.set_ylim(0, 120)

# Ajustando layout
plt.tight_layout()
plt.show()
```

Interepretação do gráfico:

*   Com os gráficos de radar chart, podemos perceber que alguns status possuem médias específicas enquanto outros são mais balanceados
*   Os tipos Grass, Water, Poison, Bug e Ice são mais balanceados
*   Os tipos Electric, Fire, Dark, Psychic, Dragon e Fighting são mais ofensivos (fisicamente ou especialmente)
*   Os tipos Steel e Rock possuem boas combinações de status ofensivos e defensivos
*   O tipo Normal possui bastante HP
*   Fairy possui uma combinação boa do lado especial (ataque e defesa) enquanto o tipo Ground do lado físico (e HP)
*   Ghost apresenta bons ataques e defesas no geral
*   Flying possui bastante speed
