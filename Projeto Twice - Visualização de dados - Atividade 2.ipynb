# Twice

Twice é um grupo sul-coreano de K-pop formado pela JYP Entertainment em 2015. O grupo é conhecido por sua música cativante, coreografias energéticas e conceitos coloridos. Ele é composto por nove integrantes, cada uma com papéis distintos como vocalistas, dançarinas e rappers.

Principais pontos sobre o Twice:

*   Formação e Popularidade: Twice foi formado através do programa de sobrevivência
"Sixteen". Desde então, o grupo se tornou um dos nomes mais populares e influentes no K-pop, tanto na Coreia do Sul quanto internacionalmente.

*   Estilo Musical: O grupo é conhecido por suas músicas pop vibrantes e dançantes, frequentemente com letras e conceitos que promovem uma imagem jovem e alegre.

*   Integrantes: São nove membros com diversos papéis como vocalistas, dançarinas e rappers. Cada uma delas traz algo único ao grupo, seja na dança, no rap ou no canto.

*   Impacto e Sucesso: Twice tem uma grande base de fãs e acumulou numerosos prêmios e recordes, incluindo vendas de álbuns e visualizações de vídeos. Elas também têm uma forte presença nas redes sociais e mídia.

*   Show no Brasil: Em fevereiro de 2024 o grupo veio ao Brasil pela primeira vez registrando um público de aproximadamente 66 mil pessoas.

Posições no grupo:

*   Nayeon: Lead Vocalist e Center

*   Jeongyeon: Lead Vocalist

*   Momo: Main Dancer, Sub-Rapper e Sub Vocalist

*   Sana: Sub Vocalist

*   Jihyo: Líder e Main Vocalist

*   Mina: Main Dancer e Sub Vocalist

*   Dahyun: Lead Rapper e Sub Vocalist

*   Chaeyoung: Main Rapper e Sub Vocalist

*   Tzuyu: Lead Dancer, Sub Vocalist, Visual e Maknae

Explicação de termos sobre os membros:

1. Lead Vocalist: É o membro que canta a maior parte das partes principais das músicas, mas não é o principal vocalista. Tem um papel importante em apresentar o vocal principal em muitas das músicas do grupo.

2. Main Vocalist: É o vocalista principal do grupo. Normalmente, é o membro com a voz mais forte e técnica, responsável pelas partes mais desafiadoras e importantes das músicas.

3. Main Rapper: É o principal rapper do grupo, responsável pelas partes de rap mais significativas e complexas das músicas.

4. Lead Rapper: É o rapper secundário que ainda tem um papel importante nas partes de rap, mas não é o principal.

5. Main Dancer: É o dançarino principal do grupo, muitas vezes responsável pelas coreografias mais complexas e impressionantes.

6. Lead Dancer: É o dançarino secundário, que ainda desempenha um papel importante nas danças, mas não é o principal.

7. Sub Vocalist: É o vocalista que fornece suporte vocálico em partes secundárias das músicas, ajudando a preencher os espaços e harmonias.

8. Sub Rapper: É o rapper que dá suporte ao rapper principal e secundário, geralmente assumindo papéis menores em partes de rap.

9. Visual: É o membro considerado o mais atraente do grupo, frequentemente destacado em termos de aparência.

10. Maknae: É o membro mais jovem do grupo.

11. Center: É o membro que geralmente ocupa a posição central durante as performances, o que ajuda a chamar a atenção do público.

# Gráfico de barra: line distribution por membro

```python
# Lista de vocais únicos
vocals = twice_lines_individual['vocal1'].unique()

# Criando um gráfico para cada vocal
for vocal in vocals:
    # Filtrando o DataFrame para o vocal atual
    vocal_df = twice_lines_individual[twice_lines_individual['vocal1'] == vocal]

    # Ordenando pelo music_order
    vocal_df = vocal_df.sort_values(by='music_order')

    # Criando o gráfico de barras
    fig = px.bar(
        vocal_df,
        x='song_name',
        y='duration',
        title=f'Tempo cantado de cada música: {vocal}',
        labels={'song_name': 'Song Name', 'duration': 'Duration'},
        text='duration'
    )

    # Ajustando o layout
    fig.update_layout(
        xaxis_title='Nome da Música',
        yaxis_title='Duração (s)',
        xaxis_tickangle=-45,
        yaxis=dict(range=[0, 60])
    )

    # Mostrando o gráfico
    fig.show()
```

# Pie chart: line distribution de Talk That Talk (música mais recente)

```python
# Filtrando o dataframe para a música "Talk That Talk"
talk_that_talk_df = twice_lines_individual[twice_lines_individual['song_name'] == 'TalkThatTalk']

# Calculando o total de duração da música
total_duration = talk_that_talk_df['duration'].sum()

# Calculando o percentual de duração para cada vocal
talk_that_talk_df['percentage'] = (talk_that_talk_df['duration'] / total_duration) * 100

# Criar o gráfico de pizza
fig = px.pie(
    talk_that_talk_df,
    names='vocal1',
    values='percentage',
    title='Percentual de Duração de Cada Membro em "Talk That Talk"',
    labels={'vocal1': 'Integrante', 'percentage': 'Percentual'}
)

# Mostrar o gráfico
fig.show()
```

Com o Pie Chart podemos ver que as maiores durações estão distribuídas entre as principais vocalistas (Nayeon e Jihyo).
A terceira vocalista, Jeongyeon, nessa época, estava passando por questões de saúde e não estava com todas as atividades de sempre no grupo. Isso explica a baixa na quantidade de linhas.

# Nuvem de palavras: palavras mais comuns nas músicas

```python
# Baixando os recursos do NLTK
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# String contendo todo o texto da coluna 'lyric'
text = ' '.join(twice_lines['lyric'].astype(str))

# Tokenizando o texto
words = word_tokenize(text.lower())

# Stopwords para inglês e coreano
stop_words_english = set(stopwords.words('english'))
stop_words_korean = set(['의', '가', '이', '은', '들', '는', '하다', '들', '에', '가', '으로', '에서', '의'])
stop_words = stop_words_english.union(stop_words_korean)
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]

# Contagem da frequência das palavras
word_freq = pd.Series(filtered_words).value_counts().reset_index()
word_freq.columns = ['word', 'frequency']
```

```python
# Criando a nuvem de palavras
fig = px.treemap(word_freq,
                 path=['word'],
                 values='frequency',
                 title='Nuvem de Palavras')
fig.show()
```

Palavras que mais se repetem nas músicas:

knock - onomatopeia de batida (todas vindo da mesma música: Knock Knock)

nae - não

deo - mais

baby - forma amorosa de chamar alguém

likey - curtir (todas vindo da mesma música: Likey)

nal - dia

yeah - interjeição

nan - eu

bonae - mandar, enviar (todas vindo da mesma música: Signal)

naega - eu

Conclui-se que o grupo tem como recorrência em suas composições mais antigas a repetição de uma palavra ao longo de toda a música.

# Rank de música

```python
# Filtrando apenas as músicas do Twice
twice_df = kpop_rankings[kpop_rankings['artist'] == 'TWICE']

# Agrupando por título da música
twice_rankings = twice_df.groupby('song_title').agg({'rank': 'min', 'year': 'first', 'time': 'first'}).reset_index()

# Ordenando pela posição mais alta
twice_rankings_sorted = twice_rankings.sort_values(by='rank')

# Filtrando apenas as músicas que pegaram TOP 50
twice_rankings_sorted_filtered = twice_rankings_sorted[twice_rankings_sorted['rank'] <= 50]
```

```python
# Criando o gráfico de barras horizontais
fig = px.bar(twice_rankings_sorted_filtered,
             x='rank',
             y='song_title',
             text='rank',
             title='Rank mais altos do Twice',
             labels={'rank': 'Rank mais alto', 'song_title': 'Música'},
             color='rank',
             color_continuous_scale=px.colors.sequential.Plasma)

# Adicionando anotações para ano e horário
for index, row in twice_rankings_sorted_filtered.iterrows():
    fig.add_annotation(
        x=row['rank'] + 0.5,
        y=row['song_title'],
        text=f"{row['year']} {row['time']}",
        showarrow=False,
        font=dict(size=10, color="black")
    )

fig.update_layout(
    yaxis_title='Nome da Música',
    xaxis_title='Rank mais altos do Twice',
    yaxis=dict(autorange='reversed')
)

fig.show()
```

# Dispersão: rank mais alto de cada música ao longo dos anos

```python
# Criar uma coluna adicional para o tamanho, invertendo a ordem do ranking
twice_rankings_sorted['size'] = 200 - twice_rankings_sorted['rank']

# Criando o gráfico de dispersão
fig = px.scatter(twice_rankings_sorted,
                 x='year',
                 y='rank',
                 size='size',
                 color='rank',
                 color_continuous_scale=px.colors.sequential.Plasma,
                 title='Dispersão dos Rankings das Músicas do Twice',
                 labels={'rank': 'Ranking', 'year': 'Ano'})

# Atualizando layout para inverter a ordem do eixo y
fig.update_layout(
    yaxis_title='Ranking',
    xaxis_title='Ano',
    yaxis=dict(autorange='reversed')
)

fig.show()
```
